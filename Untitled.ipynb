```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify

df = pd.read_csv('dataset_olympics.csv')

df.info()
```

```python
df_cleaned = df.dropna(subset=['Age', 'Medal'])

bins = [0, 15, 20, 25, 30, 35, 40, 45, 50, 100]
labels = ['15 and under', '16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51 and over']
df_cleaned['Age Group'] = pd.cut(df_cleaned['Age'], bins=bins, labels=labels, right=False)

medal_count_by_age = df_cleaned.pivot_table(index='Age Group', columns='Medal', aggfunc='size', fill_value=0)

medal_count_by_age = medal_count_by_age[['Gold', 'Silver', 'Bronze']]

plt.figure(figsize=(14,7))
medal_count_by_age.plot(kind='bar', stacked=True)
plt.title('Medal Counts by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Medals')
plt.legend(title='Medal Type')

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#pretty much expected, although suprising there is some older winners. I wonder what sports they were competing in..
```

```python

older_threshold = 35

older_gold_medalists = df_cleaned[(df_cleaned['Age'] > older_threshold) & (df_cleaned['Medal'] == 'Gold')]

gold_medals_by_sport = older_gold_medalists['Sport'].value_counts()

plt.figure(figsize=(14,10))
squarify.plot(sizes=gold_medals_by_sport, label=gold_medals_by_sport.index, alpha=0.8, color=plt.cm.tab20.colors)
plt.title('Gold Medals Won by Older Athletes (Over 35) by Sport')
plt.axis('off')
plt.show()
```

```python
#less physically demanding competitions
```

```python
#Let's see how countries have performed over time:

country_performance = df_cleaned.pivot_table(index='Year', columns='NOC', aggfunc='size', fill_value=0)

top_countries = country_performance.sum().sort_values(ascending=False).head(10).index
plt.figure(figsize=(14,7))
country_performance[top_countries].plot()
plt.title('Country Performance Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.legend(title='Country', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```
